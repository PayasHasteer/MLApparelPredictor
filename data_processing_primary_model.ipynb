**Function: Get list of images from each category**

```python
import requests 
from bs4 import BeautifulSoup 
import urllib.request
```

```python
list_of_categories = [
"mens+tshirt",
"mens+shorts",
"mens+suit+jacket",
"dress",
"high+heels",
"sunglasses",
"watches",
"socks",
]

dict_of_categories = {
"mens+tshirt": 46,
"mens+shorts": 15,
"mens+suit+jacket": 5,
"dress": 59,
"high+heels": 7,
"sunglasses": 80,
"watches": 33,
"socks": 17,
}
```

```python
list_of_gen_links = []


for item in list_of_categories:
  list_of_gen_links.append(f"https://www.myer.com.au/search?query={item}&pageNumber=")

print(list_of_gen_links)
```

```python
def getdata(url): 
    r = requests.get(url) 
    return r.text 

def get_list_image_websites(url):
  #"""Function to obtain the links for each of the images with the specified url"""
  htmldata = getdata(url)
  soup = BeautifulSoup(htmldata, 'html.parser') 

  image_url_list = []
  for item in soup.find_all('img'):
    image_url_list.append(item['src'])
  
  return image_url_list
```

```python
# place lists of links of websites into the array
categorical_links = []

for index, item in enumerate(list_of_gen_links):
  the_added_list = []
  num_pages = dict_of_categories[list_of_categories[index]]
  for i in range(1, num_pages + 1):
    list_of_items = get_list_image_websites(item + str(i))
    for innerItem in list_of_items:
      if "16x20" in innerItem or '/static' in innerItem:
        continue
      else:
        the_added_list.append(innerItem)

  categorical_links.append(the_added_list)
  print(the_added_list)

print(categorical_links)
```
