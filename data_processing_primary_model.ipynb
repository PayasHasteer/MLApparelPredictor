````python
import requests 
from bs4 import BeautifulSoup 
import urllib.request
````

````python
list_of_categories = [
"mens+tshirt",
"mens+shorts",
"mens+suit+jacket",
"dress",
"high+heels",
"sunglasses",
"watches",
"socks",
]

dict_of_categories = {
"mens+tshirt": 46,
"mens+shorts": 15,
"mens+suit+jacket": 5,
"dress": 59,
"high+heels": 7,
"sunglasses": 80,
"watches": 33,
"socks": 17,
}
#hello
````

````python
list_of_gen_links = []


for item in list_of_categories:
  list_of_gen_links.append(f"https://www.myer.com.au/search?query={item}&pageNumber=")

print(list_of_gen_links)
````

````python
def getdata(url): 
    r = requests.get(url) 
    return r.text 

def get_list_image_websites(url):
  #"""Function to obtain the links for each of the images with the specified url"""
  htmldata = getdata(url)
  soup = BeautifulSoup(htmldata, 'html.parser') 

  image_url_list = []
  for item in soup.find_all('img'):
    image_url_list.append(item['src'])
  
  return image_url_list
````

````python
# place lists of links of websites into the array
categorical_links = []

for index, item in enumerate(list_of_gen_links):
  the_added_list = []
  num_pages = dict_of_categories[list_of_categories[index]]
  for i in range(1, num_pages + 1):
    list_of_items = get_list_image_websites(item + str(i))
    for innerItem in list_of_items:
      if "16x20" in innerItem or '/static' in innerItem:
        continue
      else:
        the_added_list.append(innerItem)

  categorical_links.append(the_added_list)
  print(the_added_list)

print(categorical_links)
````

````python
for item in categorical_links:
  print(item)
````

````python
def formulate_dir_for_GD_mount(category, index, data_type):
  string_to_return = "/content/gdrive/My Drive/images/MyersSet/" + data_type + "/" + category + "/" + "myers_" + category + "_" + str(index) + ".jpg"

  return string_to_return
````

````python
print(formulate_dir_for_GD_mount("hello", 0, "training"))
````

````python
from urllib.error import HTTPError
````

````python
index = 0
data_type = ""

for outer_array in categorical_links:
  innerIndex = 0
  sixty_len = int(0.6*len(outer_array))
  twenty_len = int(0.2*len(outer_array))
  for item in outer_array:
    # seperate the links into a 60 20 20 split

    if innerIndex < sixty_len:
      data_type = "training"
    if innerIndex > sixty_len and innerIndex < sixty_len + twenty_len:
      data_type = "validation"
    if innerIndex > sixty_len + twenty_len:
      data_type = "testing"

    print(formulate_dir_for_GD_mount(list_of_categories[index], innerIndex, data_type ))

    try:
      urllib.request.urlretrieve( item, formulate_dir_for_GD_mount(list_of_categories[index], innerIndex, data_type ))
    except HTTPError as e:
      continue

    innerIndex += 1
  index += 1
````
